```python
def userInputInt(msg: str):
    var = input(msg)
    if var == "":
        return 0
    else:
        try:
            # 또는 int(value)로 정수만 확인할 수 있습니다.
            return int(var)
        except ValueError:
            return False

    # if not str.isdigit(var):
    #     return 0
    # else:
    #     return int(var)


def PrintArray(list: list):
    print()
    i: int = 0
    for item in list:
        print(f"{i} : {item}")
        i += 1
```
